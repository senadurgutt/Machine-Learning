### Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np
```

### Machine Learning Lessons

```python
name = []
mileage = []
dealer_name = []
rating = []
review_count = []
price = []

for i in range (1,13):
    
    # website in variable
    website = 'https://www.cars.com/shopping/results/?page='+ str(i) +'&page_size=20&dealer_id=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=20&mileage_max=&page_size=80&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip=' 
    
    # request to website
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('div', {'class' : 'vehicle-card'})
    
    # loop through results
    for result in results:
    
        # name
        try:
            name.append(result.find('h2').get_text()) 
        except:
            name.append('n/a')

        # mileage
        try:
            mileage_text = result.find('div', {'class': 'mileage'}).get_text()
            cleaned_mileage = mileage_text.replace('mi.', '').replace(',', '').strip()
            mileage.append(cleaned_mileage)
        except:
            mileage.append('n/a')

        # dealer_name
        try:
            dealer_name.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')

        # rating
        try:
            rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
            
        except:
            rating.append('n/a')

        # review_count
        try:
            raw_review_count = result.find('span', {'class':'sds-rating__link'}).get_text()
            cleaned_review_count = raw_review_count.replace(',', '').strip()
            review_count.append(cleaned_review_count)

            
        except:
            review_count.append('Null')

        #price 
        try:
            raw_price = result.find('span', {'class':'primary-price'}).get_text()
            price.append(raw_price.replace('$', '').replace(',', '').strip())
        except:
            price.append('n/a')
```

```python
# dictionary
car_dealer = pd.DataFrame({'Name': name, 'Mileage':mileage, 'Dealer_Name':dealer_name,
                                'Rating': rating, 'Review_Count': review_count, 'Price': price})
```

```python
car_dealer['Review_Count'] = car_dealer['Review_Count'].apply(lambda x: x.strip('reviews)').strip('('))
```
