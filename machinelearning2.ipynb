```python
def single_distplot(column):
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.distplot(df[column], bins=40)
    plt.title(f'Distribution of {column}')
    plt.show()

single_distplot('Price')
```

```python
missing_values = df.isna().sum()

# Sonuçları göster
print(missing_values)
```

```python
df.dropna(axis=0, inplace=True)
print(df.isna().sum())
```

```python
df["Name"] = df.Name.str[:+13]
```

```python
nan_rows = df[df["Review_Count"].isna()]
print(nan_rows)
```

```python
df.drop('Dealer_Name', axis=1, inplace=True)
```

```python
labelEncoder = LabelEncoder()
print(df["Name"].value_counts())
df["Name"] = labelEncoder.fit_transform(df["Name"])
print(df["Name"].value_counts())
df.describe()
```

```python
labelEncoder = LabelEncoder()
print(df["Review_Count"].value_counts())
df["Review_Count"] = labelEncoder.fit_transform(df["Review_Count"])
print(df["Review_Count"].value_counts())
df.describe()
```
